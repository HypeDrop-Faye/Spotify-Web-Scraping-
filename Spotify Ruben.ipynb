# Fonction Nettoyage

```python
def nettoyage(df):
    
    global df_final
    df_final = df.copy()
    
    for y, el in enumerate (df_final['titre']):
        
        try:
            position = el.find(',')
            df_final['titre'][y] = el[: position]
            
        except:
            
            pass
  
    for el in df_final.columns:
        
        for s in range(0, len(df_final)):
            
            try:

                df_final[el][s] = df_final[el][s].replace('likes' , '').replace('like' , '').replace('\u202f', '').replace('&nbsp;' , '').replace('&amp', '') \
            .replace('<span class="_0462465b64c9df5f31f2601f892b57d3-scss">',' ').replace('</span>', ' ').replace('playlists publiques', '').replace('abonn√©s','').replace('abonn√©','') \
            .replace('titres','').replace('playlist publique', '').replace('titre','')
        
            except:
                
                pass
            
    for u , el in enumerate (df_final.columns):
        
        try:
            
            df_final[el] = df_final[el].str.strip()
            df_final[el][u] = re.sub(' +', ' ', df_final[el][u]) 
            #df_final[el][u] = el.encode('utf-8').decode('utf-8', 'ignore')
        
        except:
            
            pass
    try:
        
        df_final['nom_artiste'] = df_final['nom_artiste'].astype(str)
        df_final['nom_personne'] = df_final['nom_personne'].astype(str)
        df_final['nom_de_la_playlist'] = df_final['nom_de_la_playlist'].astype(str)
        df_final['duree'] = df_final['duree'].astype(str)

        var_num = ['like', 'nbr_playlist_personne' , 'nbr_followers', 'titre' ]
        
    except:
        
        pass
    
    for ele in var_num:
        
        try:
        
            df_final[ele] = df_final[ele].str.replace(" ", "")
       
        except:
            
            pass
        
        for z in range(0, len(df_final)):
            
            try:
                
                df_final[ele][z] = pd.to_numeric(df_final[ele][z])
                
            except:
                
                pass
    
    return df_final
```

# Param√©trage Proxy

```python
def prox():
       
    
    async def show(proxies):
        
        global proxy_list
        proxy_list = []
        
        while True:
            proxy = await proxies.get()
            if proxy is None: break
            proxy_list.append(proxy)
            print('Found proxy: %s' % proxy)

    proxies = asyncio.Queue()
    broker = Broker(proxies)
    tasks = asyncio.gather(
        broker.find(types=['HTTP', 'HTTPS'], lvl=['High'] , countries= ['GB' ,'CA' , 'US'] ,  strict = True, port= ['8888'], host=['127.0.0.1'] , limit=10),  #, countries= ['FR'] , port= ['8888'], lvl=['Anonymous'], 
        show(proxies))

    loop = asyncio.get_event_loop()
    broker._resolver = resolvercustom.ResolverCustom(loop=loop)
    

    return loop.run_until_complete(tasks)
```

```python
def nettoyage_prox(proxy_list):
    
    global propre_proxy
    propre_proxy = []
    
    for el in proxy_list:
        el = str(el)
        deb = el.find(']')+1
        pp = el[deb:-1].replace(' ', '')
        propre_proxy.append(pp)
        
    return propre_proxy
```

```python
def fireprox(proxy_n):
    
    global firefox_capabilities,options
    
    options = Options()
    options.add_argument("user-agent=%s" %ua.random)  
    
    firefox_capabilities = webdriver.DesiredCapabilities.FIREFOX
    firefox_capabilities['marionette'] = True

    firefox_capabilities['proxy'] = {
    "proxyType": "MANUAL",
    "httpProxy": proxy_n,
    "ftpProxy": proxy_n,
    "sslProxy": proxy_n
}

   # return 
```

```python
def test_proxy(propre_proxy):
    
    global propre_proxy2, page_annonce, blaze, prox
    propre_proxy2 = []

    for prox in propre_proxy:
        
        fireprox(prox)
        
        try :
            
            get_url(page_annonce)
            time.sleep(5)
           # playlists2(blaze)
            driver.delete_all_cookies() 
            driver.quit()
            propre_proxy2.append(prox)
            
            
        except: 
            
            driver.delete_all_cookies() 
            driver.quit()
            pass 
        
     
    return propre_proxy2
```

```python
def param_proxy():
    
    global proxy_list, propre_proxy, propre_proxy2, proxy_pool, proxy_n
    
    prox()
    nettoyage_prox(proxy_list)
    test_proxy(propre_proxy)
    proxy_pool = it.cycle(propre_proxy2)
    proxy_n = next(proxy_pool)
    fireprox(proxy_n)   
```
